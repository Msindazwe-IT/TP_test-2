```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
informal = pd.read_csv("C:/Users/User/Downloads/barragan.csv")
```

```python
plt.scatter("structure_count", "ward_id")
plt.title("Scatter Plot of Structure count vs. Ward id")
plt.xlabel('toilets_total')
plt.ylabel('toilets_full_flush')
plt.show()
```

```python
Owners = informal["pocket_code"].tail()
Toilets =  [n for n in informal["structure_count"].tail()]
# Creating bar chart
plt.figure(figsize=(8, 5))
plt.bar(Owners, Toilets, color='blue', alpha=0.7)
plt.title('Toilets delivered by owners')
plt.xlabel('Owners')
plt.ylabel('Number of Toilets')
plt.show()
```

```python
market_share = [n for n in informal["ward_id"].tail()]
brands = informal["age_of_is"].tail()

plt.pie(market_share, labels=brands, autopct='%1.1f%%', startangle=140)
plt.title('informal settlement pie chart')
plt.show()
```

```python
sns.distplot(informal['ward_id'])
```

```python
sns.boxplot(x='age_of_is', y='structure_count', data=informal)
```

```python
structre_count = informal["structure_count"].head()
test_scores = np.append(structre_count, [40, 40, 90, 100, 100, 100]) 
```

```python
plt.figure(figsize=(6, 8))
plt.boxplot(test_scores, vert=True, patch_artist=True, meanline=True, showmeans=True)
plt.title('Distribution of Toilets')
plt.ylabel('Test Scores')
plt.grid(True)
plt.show()
```
